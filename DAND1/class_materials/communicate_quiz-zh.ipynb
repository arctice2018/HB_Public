```python
# 导入并加载数据
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
%matplotlib inline
df = pd.read_csv("store_data.csv")
```

```python
# 探索数据
df.head()
df.tail()
df.info()
df.describe()
df.duplicated().sum()
df.isnull().any().sum()
```

```python
# 上个月的销售额
df["week"]=pd.to_datetime(df["week"])
df.info()
```

```python
# 平均销售额
df.mean().plot(kind="pie")
```

```python
# 2016 年 3 月 13 日所在的那一周的销售额
df[df["week"]=="2016-03-13"].iloc[:,1:].plot(kind="bar")
```

```python
# 过去 3 个月的销售额
df["week"].max()
df[df["week"]>="2017-12-01"].iloc[:,1:].sum().plot(kind="pie")
```
