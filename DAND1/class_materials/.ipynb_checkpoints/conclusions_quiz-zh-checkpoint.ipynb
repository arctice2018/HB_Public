在下面的空白处探索 `store_data.csv`，回答以下测试题。

```python
# 导入并加载数据
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("store_data.csv")
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
# 探索数据
df.head()
df.info()
df.describe()
```

```python
# 首先，先清洗数据，将错误的数据类型改正确。
df["week"]=pd.to_datetime(df["week"])
# 查看更改的结果
df.info()
```

```python
# 最后一个月的总销售额
df.head()
df.tail(20)
# df.describe()
```

```python
df.hist(figsize=(8,8))
```

```python
# 平均销售额
df.mean()
```

```python
# 2016 年 3 月 13 日的销售额
df[df["week"]=="2016-03-13"]
```

```python
# C 店销售额最低的一周
df[df["storeC"]==df["storeC"].min()]
```

```python
# 最近 3 个月的总销售额
# 按照时间对原有的数据进行排序
df.sort_values("week",inplace=True)
df.head(10)
```

```python
# 最近 3 个月的总销售额
# 1：是将"week"列排除
# df.iloc[187:,1:].sum()
df[df["week"]>="2017-12-01"].iloc[:,1:].sum()
```
